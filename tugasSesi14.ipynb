```python
mahasiswa_list = []

class NIMException(Exception):
    pass

class NamaException(Exception):
    pass

class AlamatException(Exception):
    pass

class KelasException(Exception):
    pass

def menu_utama():
    print(''' 
=== MENU UTAMA ===
1. tambah data siswa
2. tampilkan data
3. exit
''')
def input_nim():
    while True:
        nim = input("Masukkan NIM : ")
        try:
            if not (len(nim) == 11 and nim.isdigit()):
                raise NIMException("NIM harus 11 digit angka!")
            return nim
        except NIMException as e:
            print(e)

def input_nama():
    while True:
        nama = input("Masukkan Nama : ")
        try:
            if len(nama.strip()) < 4:
                raise NamaException("Nama minimal 4 karakter!")
            return nama
        except NamaException as e:
            print(e)

def input_alamat():
    while True:
        alamat = input("Masukkan Alamat : ")
        try:
            if len(alamat.strip()) < 6:
                raise AlamatException("Alamat minimal 6 karakter!")
            return alamat
        except AlamatException as e:
            print(e)

def input_kelas():
    kelas_valid = ["TI24H", "TI24A", "TI24G", "TI24F"]
    while True:
        kelas = input("Masukkan Kelas (TI24H/TI24A/TI24G/TI24F): ")
        try:
            if kelas not in kelas_valid:
                raise KelasException("Kelas hanya boleh TI24H, TI24A, TI24G, TI24F!")
            return kelas
        except KelasException as e:
            print(e)

def tambah_mahasiswa():
    print("=== Tambah Data Mahasiswa ===")
    nim = input_nim()
    nama = input_nama()
    alamat = input_alamat()
    kelas = input_kelas()
    mahasiswa = {
        "nim": nim,
        "nama": nama,
        "alamat": alamat,
        "kelas": kelas
    }
    mahasiswa_list.append(mahasiswa)
    print("Data mahasiswa berhasil ditambahkan!")

def tampilkan_mahasiswa():
    print("=== Daftar Mahasiswa ===")
    if not mahasiswa_list:
        print("Belum ada data mahasiswa.")
    else:
        for i, mhs in enumerate(mahasiswa_list, 1):
            print(f"{i}. NIM: {mhs['nim']}, Nama: {mhs['nama']}, Alamat: {mhs['alamat']}, Kelas: {mhs['kelas']}")


while True:
    menu_utama()
    pilihmenu = input("Isikan pilihan menu : ")
    if pilihmenu == "1":
        tambah_mahasiswa()
    elif pilihmenu == "2":
        tampilkan_mahasiswa()
    elif pilihmenu == "3":
        print("Terima kasih!")
        break
    else:
        print("Pilihan tidak valid!")
```
